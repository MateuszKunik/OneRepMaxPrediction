```python
import numpy as np
import pandas as pd

import preprocessing_utils as pu
```

```python
# Read personal data from excel
personal_data = pd.read_excel('../../data/PersonalData.xlsx')
personal_data = pu.calculate_BMI(personal_data)
```

```python
# Read data from csv
dataframe = pd.read_csv('../../data/processed/ConvertedSquats.csv')
data = pd.merge(dataframe, personal_data, on='Id')

# Preprocessing
data = pu.prepare_regression_target(data)
data = pu.drop_irrelevant_columns(data)
data = pu.transform_features(data)
data = pu.smooth_coordinate_data(data, frac=0.1, it=3)
data = pu.reorder_dataframe_columns(data)
```
