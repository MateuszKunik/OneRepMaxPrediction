```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img1 = cv.imread('../data/test/img1.jpg')

plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
img2 = cv.imread('../data/test/img2.jpg')

plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
patternSize = (9, 6)

ret1, corners1 = cv.findChessboardCorners(img1, patternSize)
ret2, corners2 = cv.findChessboardCorners(img2, patternSize)
```

```python
corners1.shape
```

```python
plt.scatter(corners1[:, 0, 0], corners1[:, 0, 1])
```

```python
plt.scatter(corners2[:, 0, 0], corners2[:, 0, 1])
```

```python
H, _ = cv.findHomography(corners1, corners2)
print(H)
```

```python
img1_warp = cv.warpPerspective(img1, H, (img1.shape[1], img1.shape[0]))

plt.imshow(cv.cvtColor(img1_warp, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
